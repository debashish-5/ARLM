## Importing all libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.ar_model import AutoReg
```

## Creating time series forcasting

```python
data = [120,130,128,140,150,160,170,180,175,185]
ts = pd.Series(data)
#this is time series data (order by time)
```

## Train the ARLM model

```python
model = AutoReg(ts,lags = 2)
model_fit = model.fit()
#This tell the model : Look at last 2 values and learn how they influenced the last two valuee
```

## view model summery

```python
print(model_fit.summary())
```

## predict future values

```python
predictions = model_fit.predict(start=len(ts),end = len(ts)+2,dynamic = False)
#forcasts next 3 time steps
```

## Plot result

```python
plt.figure(figsize=(13,7))
plt.plot(ts,label = 'Orginal Data')
plt.plot(predictions,label ='Forecast',color='red')
plt.legend()
plt.show()
```
